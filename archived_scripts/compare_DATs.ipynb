```python
from lib.file.DAT import DATArrayFile
import matplotlib.pyplot as plt
import numpy as np
```

```python
datadir = "./tsm_targets/06-06-22/Kate_old_rig/"
filename1 = datadir + "02-01-02_snr.txt"
filename2 = datadir + "SNR02_01_02.dat"
```

```python
da1 = DATArrayFile(filename1)
da1.normalize()
da1 = da1.get_data()
da2 = DATArrayFile(filename2)
da2.normalize()
da2 = da2.get_data()

print(da1.shape, da2.shape)
print(np.max(da1), np.max(da2))
print(np.min(da1), np.min(da2))
print(da1, da2)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(da1, cmap='jet')
axes[1].imshow(da2, cmap='jet')
plt.show()
```

```python
# compare SNR px distribution
colors=["blue", 'red']
label=["Kate's PhotoZ", "John's PhotoZ"]
da = np.concatenate([da1.reshape(-1,1), da2.reshape(-1,1)], axis=1)
print(da.shape)
plt.hist(da1.reshape(-1), 100, histtype='step', stacked=True, fill=False, color=colors[0], label=label[0])
plt.hist(da2.reshape(-1), 100, histtype='step', stacked=True, fill=False, color=colors[1], label=label[1])
plt.title("SNR Distribution over Pixels")
plt.legend()
```
